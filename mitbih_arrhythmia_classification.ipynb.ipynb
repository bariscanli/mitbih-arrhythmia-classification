```python
class_1 = df[df[187]==1.0]
class_2 = df[df[187]==2.0]
class_3 = df[df[187]==3.0]
class_4 = df[df[187]==4.0]
class_0 = df[df[187]==0.0].sample(n = 8000)
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(new_df.drop([187], axis=1), new_df[187], test_size=0.15, random_state=42)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.1765, random_state=42)  
```

```python
print("Eğitim seti boyutu:", X_train.shape[0])
print("Doğrulama (Validation) seti boyutu:", X_val.shape[0])
print("Test seti boyutu:", X_test.shape[0])
```

```python
X_train = np.array(X_train).reshape(X_train.shape[0], X_train.shape[1], 1)
X_val = np.array(X_val).reshape(X_val.shape[0], X_val.shape[1], 1)
X_test = np.array(X_test).reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
print("X_train şekli:", X_train.shape)
print("X_val şekli:", X_val.shape)
print("X_test şekli:", X_test.shape)
```

```python
from tensorflow.keras import Sequential,utils
from tensorflow.keras.layers import Flatten , Dense, Conv1D, MaxPool1D, Dropout, Input
```

```python

model = Sequential([ 
    Input(shape=(X_train.shape[1], 1)), 
    Conv1D(filters=64, kernel_size=3, activation='relu'), 
    MaxPool1D(pool_size=2),
    Conv1D(filters=128, kernel_size=3, activation='relu'),
    MaxPool1D(pool_size=2),
    Conv1D(filters=256, kernel_size=3, activation='relu'),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5), 
    Dense(5, activation='softmax')  
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) #Adam optimizasyonu
```

```python
model.summary()
```

```python
history = model.fit(X_train, y_train, 
                    validation_data=(X_val, y_val),
                    epochs=20,
                    batch_size=32)
```

```python
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test Doğruluğu: {test_acc * 100:.2f}%")
```

```python
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.xlabel('Epoch')
plt.ylabel('Kayıp')
plt.legend()
plt.show()

# Eğitim ve doğrulama doğruluğunu çiz
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.xlabel('Epoch')
plt.ylabel('Doğruluk')
plt.legend()
plt.show()
```
